```python
import requests
from bs4 import BeautifulSoup

# Function to scrape product data (book title, price, and link)
def fetch_books_data(url):
    # Set headers to mimic a real browser request
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    
    # Send GET request to the page
    response = requests.get(url, headers=headers)

    # Check if request was successful
    if response.status_code == 200:
        # Parse the page content
        soup = BeautifulSoup(response.content, "html.parser")

        # Extract the container that holds all the books
        books = soup.find_all('article', class_='product_pod')

        # Loop through each book and extract information
        for book in books:
            # Get book title
            title = book.h3.a['title']
            
            # Get book price
            price = book.find('p', class_='price_color').text
            
            # Get book link (relative link, needs to be combined with base URL)
            link = book.h3.a['href']
            book_link = f"http://books.toscrape.com/catalogue/{link}"

            # Print extracted information
            print(f"Title: {title}")
            print(f"Price: {price}")
            print(f"Link: {book_link}")
            print("-" * 50)
    else:
        print(f"Failed to retrieve the webpage. Status code: {response.status_code}")

# URL of the page to scrape (homepage for Books to Scrape)
url = 'http://books.toscrape.com/'

# Call the function to scrape and display book data
fetch_books_data(url)
```
